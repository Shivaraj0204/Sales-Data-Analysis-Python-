## Reading Updated DataFrame

```python
all_data = pd.read_csv(r"all_data.csv")
all_data.head()
```

## Drop NAN Values

```python
## To find NAN in a DataFrame
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df

all_data = all_data.dropna(how='all')
all_data
```

## Remove Columns based on Condition

```python
## Filtering Data Based on a Condition
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df.head()

all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

## Adding a Column 

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32') #Changes Data Type
all_data

#all_data['Month'] = pd.DatetimeIndex(all_data['Order Date']).month
#df['month'] = df['ArrivalDate'].dt.month
#all_data.head()
```

## Convert Columns to the Correct Data Type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data.head()
```

## Question 1) What was the best Month For Sales? How much was earned that Month?

```python
results = all_data.groupby('Month').sum()
results
```

## Visualisation Using Bar Chart

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in US ($)')
plt.xlabel('Month Number')
plt.show()
```

## Adding City Column Using Apply Method

```python
### Lets Use .apply Method

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
all_data.head()
```

## Question 2) which City has Highest Sales

```python
results_city = all_data.groupby('City').sum()
results_city
```

```python
import matplotlib.pyplot as plt

#cities = all_data['City'].unique() ---> This Command display Inncorrect Values on Graph (So we will use List Comprehension)

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities,results_city['Sales'])
plt.xticks(cities , rotation = '40',size=8,weight='bold')
plt.ylabel('Sales in US ($)')
plt.xlabel('City Name')
plt.show()
```

## Finding Duplicate Values in a DataFrame.

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df.head()
```

## Jion Values from two Rows

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df.head()
```

## Droping Rows with Duplicate Values

```python
df = df[['Order ID' , 'Grouped']].drop_duplicates()

df.head()
```

## Count Unique pairs of Combinations from a Column

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
##print(count)

for key, value in count.most_common(10):
    print(key,value)



```
